````python

def clean_llm_json(s: str) -> str:
    """
    清理LLM生成的JSON字符串，将其转换为标准JSON格式
    
    处理步骤：
    1. 去除```json```代码块包装符
    2. 提取第一个JSON对象
    3. 将单引号替换为双引号
    4. 将Python布尔值和None替换为JSON格式
    5. 去除末尾的逗号
    
    Args:
        s (str): LLM生成的可能包含非标准JSON格式的字符串
        
    Returns:
        str: 标准化的JSON字符串
    """
    # 去掉 ```json ```包裹
    s = re.sub(r"```json|", "", s)
    
    # 提取第一个 {...}
    match = re.search(r"\{.*}", s, re.S)
    if match:
        s = match.group()
        
    # 替换单引号为双引号（简单场景）
    s = s.replace("'", '"')
    
    # 替换 Python None/True/False
    s = s.replace("None", "null")
    s = s.replace("True", "true")
    s = s.replace("False", "false")
    
    # 去掉尾逗号
    s = re.sub(r",(\s*[}\]])", r"\1", s)
    
    return s

def robust_parse(s):
    """
    安全解析字符串为JSON或Python字面量
    
    解析顺序：
    1. 首先尝试直接解析为JSON
    2. 如果失败，尝试解析为Python字面量
    3. 如果还失败，先清理后再尝试JSON解析
    
    Args:
        s: 待解析的字符串
        
    Returns:
        解析后的数据结构（dict, list, str等）
        
    Note:
        该函数具有容错性，能够处理多种格式的数据解析
    """
    try:
        return json.loads(s)
    except:
        try:
            import ast
            return ast.literal_eval(s)
        except:
            s = clean_llm_json(s)
            return json.loads(s)

````

````python


class JsonStreamValueParser:
    def __init__(self, target_keys=("weakPointTitle","weakPointReason")):
        self.target_keys = set(target_keys)
        
        self.state = "idle"
        self.key_buf = ""
        self.current_key = None
        
        self.prev_char = ""
        
        #
        self.bracket_depth = 0  #
        self.in_target_array = False    #
        
    def feed(self, text):
        for ch in text:
            escaped = (self.prev_char == "\\")
            
            # =====
            if self.state == "idle":
                if ch == '"':
                    self.key_buf = ""
                    self.state = "in_key"
            
            # ======
            elif self.state == "in_key":
                if ch == '"' and not escaped:
                    self.current_key = (self.key_buf if self.key_buf in self.target_keys else None)
                    self.state = "after_key"
                else:
                    self.key_buf += ch
            
            # ======
            elif self.state == "after_key":
                if ch == '"':
                    #
                    if self.current_key:
                        self.state = "in_value"
                    else:
                        self.state = "skip_value"
                elif ch == '[':
                    #
                    if self.current_key:
                        self.in_target_array = True
                        self.bracket_depth = 1
                        self.state = "in_array"
                    else:
                        self.bracket_depth = 1
                        self.state = "skip_array"
            
            # 忽略： 和空白
            elif self.state == "in_value":
                if ch == '"' and not escaped:
                    yield "\n"
                    self.state = "idle"
                    self.current_key = None
                elif escaped:
                    escape_map = {
                            'n': '\n',
                            't': '\t',
                            'r': '\r',
                            '"': '"',
                            '\\': '\\'
                    }
                    yield escape_map.get(ch, ch)
                elif ch != '\\':
                    yield ch
            
            # ====
            elif self.state == "skip_value":
                if ch == '"' and not escaped:
                    self.state = "idle"
            
            # ======
            elif self.state == "in_array":
                if ch == '[' and not escaped:
                    self.bracket_depth += 1
                elif ch == ']' and not escaped:
                    self.bracket_depth -= 1
                    if self.bracket_depth == 0:
                        yield "\n"
                        self.state = "idle"
                        self.current_key = None
                        self.in_target_array = False
                elif ch == '"' and not escaped:
                    #
                    self.state = "in_array_string"
            
            # ===
            elif self.state == "in_array_string":
                if ch == '"' and not escaped:
                    yield ", "  #
                    self.state = "in_array"
                elif escaped:
                    escape_map = {
                            'n': '\n',
                            't': '\t',
                            'r': '\r',
                            '"': '"',
                            '\\': '\\'
                    }
                    yield escape_map.get(ch, ch)
                elif ch != '\\':
                    yield ch
            
            # ====
            elif self.state == "skip_array":
                if ch == '[' and not escaped:
                    self.bracket_depth += 1
                elif ch == ']' and not escaped:
                    self.bracket_depth -= 1
                    if self.bracket_depth == 0:
                        self.state = "idle"
            
            # 
            self.prev_char = ch if (ch == '\\' and not escaped) else ''
````

````python
class JsonStreamValueParser:
    def __init__(self, target_keys=("weakPointTitle", "weakPointReason")):
        self.target_keys = set(target_keys)

        self.stack = []
        self.state = "normal"

        self.key_buf = ""
        self.current_key = None
        self.prev_char = ""

        self.target_array_depth = 0

    def _finish_value(self):
        if self.stack:
            self.stack[-1]["state"] = "after_value"
        self.current_key = None
        self.state = "normal"

    def feed(self, text):
        escape_map = {
            "n": "\n",
            "t": "\t",
            "r": "\r",
            '"': '"',
            "\\": "\\",
            "/": "/",
        }

        for ch in text:
            escaped = self.prev_char == "\\"

            # 解析 object 的 key
            if self.state == "in_key":
                if ch == '"' and not escaped:
                    self.current_key = self.key_buf
                    self.key_buf = ""
                    self.stack[-1]["state"] = "colon"
                    self.state = "normal"
                else:
                    self.key_buf += ch

                self.prev_char = ch if ch == "\\" and not escaped else ""
                continue

            # 跳过普通字符串 value
            if self.state == "skip_string":
                if ch == '"' and not escaped:
                    self._finish_value()

                self.prev_char = ch if ch == "\\" and not escaped else ""
                continue

            # 输出目标字符串 value
            if self.state == "target_string":
                if ch == '"' and not escaped:
                    yield "\n"
                    self._finish_value()
                elif escaped:
                    yield escape_map.get(ch, ch)
                elif ch != "\\":
                    yield ch

                self.prev_char = ch if ch == "\\" and not escaped else ""
                continue

            # 输出目标数组里的字符串
            if self.state == "target_array":
                if ch == '"' and not escaped:
                    self.state = "target_array_string"
                elif ch == "[" and not escaped:
                    self.target_array_depth += 1
                elif ch == "]" and not escaped:
                    self.target_array_depth -= 1
                    if self.target_array_depth == 0:
                        yield "\n"
                        self._finish_value()

                self.prev_char = ch if ch == "\\" and not escaped else ""
                continue

            if self.state == "target_array_string":
                if ch == '"' and not escaped:
                    yield ", "
                    self.state = "target_array"
                elif escaped:
                    yield escape_map.get(ch, ch)
                elif ch != "\\":
                    yield ch

                self.prev_char = ch if ch == "\\" and not escaped else ""
                continue

            # 跳过数字、true、false、null 等非字符串值
            if self.state == "literal":
                if ch in ",]}":
                    self._finish_value()
                    # 继续让 normal 状态处理这个分隔符
                else:
                    self.prev_char = ch if ch == "\\" and not escaped else ""
                    continue

            # normal 状态
            top = self.stack[-1] if self.stack else None

            if ch.isspace():
                pass

            elif top and top.get("state") == "colon" and ch == ":":
                top["state"] = "value"

            elif ch == "{":
                self.stack.append({
                    "type": "object",
                    "state": "key"
                })
                self.current_key = None

            elif ch == "}":
                if self.stack and self.stack[-1]["type"] == "object":
                    self.stack.pop()
                    self._finish_value()

            elif ch == "[":
                if self.current_key in self.target_keys:
                    self.state = "target_array"
                    self.target_array_depth = 1
                else:
                    # 关键点：
                    # 非目标数组不能直接 skip，
                    # 因为目标 key 可能就在数组里的 object 中
                    self.stack.append({
                        "type": "array",
                        "state": "value"
                    })
                    self.current_key = None

            elif ch == "]":
                if self.stack and self.stack[-1]["type"] == "array":
                    self.stack.pop()
                    self._finish_value()

            elif ch == ",":
                top = self.stack[-1] if self.stack else None
                if top:
                    if top["type"] == "object":
                        top["state"] = "key"
                    else:
                        top["state"] = "value"

                self.current_key = None

            elif ch == '"':
                top = self.stack[-1] if self.stack else None

                if top and top["type"] == "object" and top["state"] == "key":
                    self.key_buf = ""
                    self.state = "in_key"

                elif self.current_key in self.target_keys:
                    self.state = "target_string"

                else:
                    self.state = "skip_string"

            else:
                self.state = "literal"

            self.prev_char = ch if ch == "\\" and not escaped else ""
````
